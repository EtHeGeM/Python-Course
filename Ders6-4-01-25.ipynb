Her fonksiyon bir çıktı üretmek zorunda değildir, fonksiyonları bazen anlık kullanmak için script şeklinde üretiriz.

```python
# prompt: create a circle
# tanımlama - importing
import matplotlib.pyplot as plt
#head - kafa
def create_circle():
  #-------------------------------------------------------
  ## body - vücut
  # Creating the circle
  circle = plt.Circle((0, 0), 0.5, color='blue')

  # Creating the plot
  fig, ax = plt.subplots()

  # Adding the circle to the plot
  ax.add_patch(circle)

  # Setting the plot limits
  ax.set_xlim((-1, 1))
  ax.set_ylim((-1, 1))
 # --------------------------------------------------------
  # end-bitiş
  plt.show()

create_circle()
```

# Default parametreler - Önceden ayarlanmış parametreler

```python
def kumanda(kanal = "TRT",ses=None,renk=None):

  print(f"Şu anki kanal: {kanal}")

  if ses: # Eğer ses değişkeni, nonetype dan başka bir değişken değeri taşıyorsa
    print(f"Ses: {ses}")
  if renk: # Renk değişkeni, nonetype dışında bir değişken taşıyorsa
    print(f"Renk: {renk}")
```

```python
# Bir fonksiyon yazıyoruz, fonksiyon bir marketi temsil ediyor.
# Birinci parametre, müşteri mi yoksa çalışan mı olduğunu belirtiyor.
# İkinci parametre, ürün kodunu temsil ediyor.
# Üçüncü parametre, ürün satışı mı yoksa stok artışı mı olduğunu gösteriyor.

#market(kisi = "Çalışan",kod = 132, durum = "Satış")
# Kişi girilmediği zaman, ilk tercih müşteri olsun.
# kod girilmediği zaman, ürün kodu girilmedi diye bir çıktı versin.
# durum girilmediği zaman, otomatik satış olsun.
# Çıktı -> Çalışan, 132 kodlu ülker gofret adlı ürünü 4TL fiyatından sattı.
```

```python
# ürün kodlarıyla birlikte, ismi ve fiyatını da listeliyoruz.
urun_kodlari = {1:{"isim":"Gofret","fiyat":8,"stok":7},
                2:{"isim":"Cips","fiyat":15,"stok":9},
                3:{"isim":"Kola","fiyat":12,"stok":5}}
#fonksiyon tanımlaması - head - kafa
def market(kisi = "musteri",kod = None, durum = "satis"):
  # kod boş ise, kullanıcıya bir hata mesajı döndürüyoruz.
  if kod==None:
    return "Ürün kodu girmediniz"
  # aksi halde...
  else:
    # Önce, kişinin müşteri olup olmadığını kontrol ediyorum.
    # Kişi çalışan ise, stok artışı yapar.
    # Kişi müşteri ise, satın alım yapar.
    if kisi == "musteri":
      if durum == "satis":
        if kod in urun_kodlari:
          print(f"{urun_kodlari[kod]['isim']} adlı ürünü {urun_kodlari[kod]['fiyat']} TL fiyatından sattı.")
          print(f"{urun_kodlari[kod]['stok']-1} kadar ürün kaldı.")
          urun_kodlari[kod]["stok"] -= 1
        else:
          return "Böyle bir ürün yok."
      elif durum == "stok artışı":
        return "Müşteri olarak ürün girişi yapamazsınız."
      else:
        return "Geçersiz bir durum girdiniz.(satis veya stok artışı)"
    elif kisi == "calisan":
      if durum == "satis":
        return "Çalışan olarak ürün satın alımı yapamazsınız."
      elif durum == "stok artışı":
        if kod in urun_kodlari:
          urun_kodlari[kod]["stok"] += 1
          print(f"{urun_kodlari[kod]['isim']} adlı ürünün stoğu {urun_kodlari[kod]['stok']} adede yükseldi.")
          print(f"Ödenen fiyat, {urun_kodlari[kod]['fiyat']/2}")
        else:
          return "Böyle bir ürün yok."
      else:
        return "Geçersiz bir durum girdiniz.(satis veya stok artışı)"
    else:
      return "Geçersiz bir kişi girdiniz.(musteri veya calisan)"
```

```python
print(market("calisan",1,"stok artışı"))
print(market("musteri",1,"satis"))
print(market("musteri",3,"stok artışı"))
print(market("calisan",2,"satis"))
```

# Class - Sınıf Sistemi - Tekrar

__init__ = initial function - başlangıç fonksiyonu

- self = class içi değerleri temsil eder
- self tanımlaması yapıldığı class için, sonradan çağırıldığında kendi değişkenlerini tanımlamasına olanak tanır.

```python
class deneme:
  def __init__(self,isim,soyisim,yas):
    self.isim = isim
    self.soyisim = soyisim
    self.yas = yas
    print("class çalıştırıldı")
  def yazdir(self):
    print(f"{self.isim} {self.soyisim} {self.yas}")
```

```python
p1 = deneme("a","b","c")
```

```python
import random
# piyango sistemi yazıyoruz
# bir numara vardır, 1000 tane içerisinden 100 taneye çıksın.
# 20 numara olsun, şanslı 5 numara olsun.
class piyango:
  # numara listesini ve şanslı numaraları tanımlıyoruz
  def __init__(self):
    self.numaralar = [x for x in range(1,21)]
    self.lucky_numbers = random.sample(range(1, 21), 5)
  # bize 1,20 arası rastgele bir sayı veriyor
  def bilet_al(self):
    print(f"Bilet numaranız: {random.randint(1, 20)}")
  # bilet numarasını sorgulamak için önce numaralar içinde kontrol ediyor,
  # sonrasında da şanslı numaralar listesi içerisinde var mı onu kontrol ediyor.
  def bilet_sorgula(self, bilet_numarasi):
   try:
      if type(bilet_numarasi) !=int:
        raise ValueError("Bilet numarası, sayı olmak zorundadır.")
      else:
        if bilet_numarasi in self.numaralar:
          if bilet_numarasi in self.lucky_numbers:
            print("Tebrikler, 100 Milyon TL kazandınız!")
          else:
            print("Üzgünüz, kazanamadınız")
        else:
          print("Böyle bir bilet numarası bulunmamaktadır.")
   except ValueError as e:
    print(e)

  # bayi içinin simüle edilmesi için bir while döngüsüyle mesajları döngüye aldıktan sonra
  # kullanıcının girdisini kontrol edip bilet alma ve sorgulama işlemlerini sağlıyoruz.
  def bayi(self):
    while True:
      tercih = input("Ne yapmak istiyorsunuz - \"Bilet Al\" - \" Bilet Sorgula \" - \" Çık \" ")
      if tercih == "Bilet Al":
        self.bilet_al()
      elif tercih == "Bilet Sorgula":
        try:
          bilet_numarasi = int(input("Bilet numaranızı giriniz: "))
          if type(bilet_numarasi) != int:
            raise ValueError("Bilet numarası, sayı olmak zorundadır.")
          else:
            self.bilet_sorgula(bilet_numarasi)
        except:
          print("Bilet numarası, sayı olmak zorundadır.")
      elif tercih == "Çık":
        print("Çıkış yapıldı")
        break
      else:
        print("Geçersiz bir seçim yaptınız")
```

```python
kumarbaz = piyango()
```

```python
kumarbaz.bayi()
```
