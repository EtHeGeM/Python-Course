```python
# prompt: create 3 sets and make a covariance and variance matrix of these sets

import numpy as np

def covariance_variance_matrix(set1, set2, set3):
    """
    Calculates the covariance and variance matrix for three sets.

    Args:
      set1: The first set of numbers.
      set2: The second set of numbers.
      set3: The third set of numbers.

    Returns:
      A tuple containing the covariance matrix and the variance matrix.
      Returns an error message if input sets are not of the same length or not numerical.
    """
    try:
        # Convert sets to numpy arrays
        arr1 = np.array(set1, dtype=float)
        arr2 = np.array(set2, dtype=float)
        arr3 = np.array(set3, dtype=float)

        if not (len(arr1) == len(arr2) == len(arr3)):
            return "Error: Input sets must have the same length."

        #Combine the arrays for easier calculation
        combined_arr = np.stack((arr1, arr2, arr3), axis=0)

        #Calculate covariance matrix
        cov_matrix = np.cov(combined_arr)

        #Calculate variance matrix (diagonal of covariance matrix)
        var_matrix = np.diag(cov_matrix)

        return cov_matrix, var_matrix

    except (ValueError, TypeError):
        return "Error: Input sets must contain only numerical values."

# Example usage:
set1 = [1, 2, 3, 4, 5]
set2 = [6, 7, 8, 9, 10]
set3 = [11, 12, 13, 14, 15]


cov_matrix, var_matrix = covariance_variance_matrix(set1, set2, set3)

if isinstance(cov_matrix, str):
    print(cov_matrix) #Print error if any.
else:
    print("Covariance Matrix:\n", cov_matrix)
    print("\nVariance Matrix:\n", var_matrix)
```

```python
class prog_training:
  def __init__(self):
    pass
  def cift_getir(self,liste):
    # ben bu class içerisinden cift_getir fonksiyonunu çağırdığımda, bir liste sunacağım.
    # Bu liste içerisindeki tüm çift sayıları farklı bir listeye atacak ve en sonunda çıktı verecek.
    # kullanıcıdan aldığım listenin içerisinde bir döngü oluşturuyorum.
    # Sonradan eleman eklemek için listeyi boş bir şekilde tanımlıyorum
    cift = []
    for eleman in liste:
      #dönen her bir elemanın çift olup olmadığını kontrol ediyorum.
      if eleman%2==0:
        # eğer eleman çift ise çift listesine ekliyorum
          cift.append(eleman)
          print(cift)
      else:
        print(f"{eleman} çift değildir.")
    # en sonunda listeyi çıktı veriyorum.
    print("Çift elemanlar:{}".format(cift))
    return cift
  def sesli_buyut(self, text):
    #kullanıcıdan aldığımız text isimli string ifadenin içerisinde dönüp,
    # sesli harf gördüğünde onları büyütüp tekrar yazdırsın
    # sesli harfler için liste oluştur, for döngüsüne al, if else uygula, büyüt, çıktı ver
    # text = "python kursunun 1. ayını doldurmuş bulunuyoruz."
    # çıktı : pythOn kUrsUnUn 1. AyInI dOldUrmUş bUlUnUyOrUz.

    #programın kontrol etmesi için bir sesli harfler dizisi yazdık
    sesli = "aeıioöuüAEIİOÖUÜ"
    #sonradan ekleme yapmak için boş bir txt oluşturduk.
    txt= ""
    #textin içerisindeki harfleri teker teker döngüye aldık
    for harf in text:
      # harf, sesli harfler dizisinde bulunuyorsa, harfi büyüttük ve yazdırdık.
      if harf in sesli:
        #boş txt ifademize harfi büyük halini ekledik.
        txt+=harf.upper()
      #sesli harf değilse de harfin normal halini ekledik.
      else:
        txt+=harf
    # son olarak txt değişkenimizi çıktı aldık.
    return txt


```

Kullanıcının birkaç sene içerisinde yaptığı harcamaların ortalamalarını alacak bir fonksiyon, bu fonksiyonun çıktılarını belirli bir formatta yazdıracak başka bir fonksiyon.

```python
class banka:
  # init fonksiyonuyla giriş bilgilerini alıyorum.
  def __init__(self):
    self.veri = {}

  # kullanıcının tc no, isim ve soyismini alarak kullanıcıya hesap açıyorum.
  def kullanici_ekle(self,tc,isim,soyisim):
    self.veri[tc] = {"isim":isim,"soyisim":soyisim}
    print(f"{isim} {soyisim} adlı kullanıcı eklenmiştir.")

  # kullanıcıların bilgileri listeliyorum.
  def kullanici_listele(self):
    for tc in self.veri:
      print(f"{tc} : {self.veri[tc]['isim']} {self.veri[tc]['soyisim']}")

  # kullanıcının tc nosu ile harcamalarını sisteme giriyorum.
  def harcama_ekle(self, tc,harcamalar):
    self.veri[tc] = {"isim":self.veri[tc]["isim"],
                     "soyisim":self.veri[tc]["soyisim"],
                     "harcama":harcamalar}
    print(f"{self.veri[tc]['isim']} adlı kullanıcının {harcamalar} harcaması kaydedilmiştir.")

  # kullanıcının harcama ortalamasını alıyorum.
  def harcama_ortalama(self, tc):
    harcama_listesi = self.veri[tc]["harcama"]
    ortalama = sum(harcama_listesi)/len(harcama_listesi)
    print(f"{self.veri[tc]['isim']} adlı kullanıcının ortalama harcaması {ortalama:.2f} TL'dir.")

  def kullanici_giris_ekrani(self):
    tc = int(input("TC numaranızı giriniz:"))
    if tc in self.veri.keys():
        if self.veri[tc]["isim"] is not None:
          print(f"Hoşgeldiniz {self.veri[tc]['isim']} {self.veri[tc]['soyisim']}")
          if self.veri[tc]["harcama"] is not None:
            self.harcama_ortalama(tc)
          else:
            harcamalar = []
            i=0
            while i<=3:
              harcama = int(input("Harcamayı giriniz:"))
              harcamalar.append(harcama)
              i+=1
            self.harcama_ekle(tc,harcamalar)
            print("Harcamalarınız eklendi, lütfen tekrar giriş yapınız.")
            kullanici_giris_ekrani()
        else:
          isim = input("İsminizi giriniz:")
          soyisim = input("Soyisminizi giriniz:")
          self.kullanici_ekle(tc,isim,soyisim)
    else:
      tc = input("Tc numaranızı giriniz:")
      isim = input("İsminizi giriniz:")
      soyisim = input("Soyisminizi giriniz:")
      harcamalar = []
      i=0
      while i<=3:
        harcama = int(input("Harcamayı giriniz:"))
        harcamalar.append(harcama)
        i+=1
      self.kullanici_ekle(tc,isim,soyisim)
      self.harcama_ekle(tc,harcamalar)
      print("Harcamalarınız eklendi, lütfen tekrar giriş yapınız.")
      self.kullanici_giris_ekrani()


```
