# Python To-Do List Örneği

```python
class TodoList:
    def __init__(self):
        self.todos = []

    def add(self, task):
        self.todos.append(task)

    def list(self):
        for idx, task in enumerate(self.todos, 1):
            print(f"{idx}. {task}")

    def remove(self, idx):
        try:
            self.todos.pop(idx - 1)
        except IndexError:
            print("Task not found.")

class CLI:
    def __init__(self):
        self.todo_list = TodoList()

    def run(self):
        while True:
            command = input("Enter command (add, list, remove, exit): ")
            if command == "exit":
                break
            elif command == "add":
                task = input("Enter the task: ")
                self.todo_list.add(task)
            elif command == "list":
                self.todo_list.list()
            elif command == "remove":
                idx = int(input("Enter task number to remove: "))
                self.todo_list.remove(idx)
            else:
                print("Unknown command!")

if __name__ == "__main__":
    cli = CLI()
    cli.run()
```

```python
class mobilya:
  def __init__(self):
    self.envanter = []

  def ekle(self,urun):
    self.envanter.append(urun)

  def cikar(self,urun):
    self.envanter.remove(urun)

  def listele(self):
    for idx,urun in enumerate(self.envanter,1):
      print(f"{idx}. {urun}")
```

# Try - Except Hata Ayıklama Yapıları

```python
try:
  sayi = input("Bir sayı giriniz:")
  print(sayi + 8)
except TypeError:
  print("Girdinin bir sayı olması gerekiyor")
```

```python
def uzunluk(girdi):
  #çalıştırılacak ifadedir
  try:
    print(f"{girdi} ifadesinin uzunluğu:{len(girdi)}")
  # hata alındığında uygulanacak ifadedir
  except TypeError:
    print("Girdiğiniz değer döndürülebilir bir değer olmalıdır")
  # hata alınmadığında uygulanacak ifadedir
  else:
    print("İşlem başarılı")
  # hata alınsın ya da alınmasın uygulanacak ifadedi
  finally:
    print("İşlem tamamlandı")
```

# Classlar ile adam asmaca

```python
# İlk olarak, bir kelime belirlenmesi gerekiyor
# random kütüphanesi kullanarak rastgele bir alacağım
import random

class adam_asmaca:
  def __init__(self,liste):
    self.kelime_listesi = liste
  def __str__(self):
    return "Adam asmaca oynamak için lütfen adam_asmaca.play() fonksiyonunu çalıştırın."
  def play(self):
    #Kullanıcı karşılama mesajı
    self.kelime = random.choice(self.kelime_listesi)
    print("Kelimeyi tahmin etmek için harf ya da kelime girin")
    print(len(self.kelime)*"_")
    #Kullanıcıdan tahmin alma
    counter = 0
    dogru_harfler = []
    yanlis_harfler = []
    while True:
      # counter değeri 7den az olmalı
      if counter >=7:
        return "Tahmin sayısını aştınız"
      # kullanıcıdan tahminini alacağım
      # eğer kullanıcının tahmini tek harf ise bu bir harf tahminidir.
      # eğer kullanıcının tahmini bir kelime ise bu bir kelime tahminidir.

      try:
        tahmin = input("Tahmininizi giriniz")
        if len(tahmin) == 1:
          if tahmin in self.kelime:
            # eğer harf kelimenin içerisinde varsa o harfi kaç kere görünüyorsa görünür yapmam gerekiyor
            dogru_harfler.append(tahmin)
            counter +=1
            print(f"\"{tahmin}\" harfi kelime içerisinde vardır.")
          else:
            yanlis_harfler.append(tahmin)
            counter +=1
            print("Yanlış tahmin")
        elif len(tahmin) == len(self.kelime):
          if tahmin == self.kelime:
            print("Tebrikler kelimeyi doğru tahmin ettiniz")
            for i in tahmin:
              print(i, end = " ")
            break
          else:
            for i in tahmin:
              if i not in self.kelime:
                yanlis_harfler.append(i)
                print(f"{i} harfi kelimede yok")
                counter +=1
            print("Yanlış tahmin")
        else:
          raise ValueError
      except ValueError:
        print("Tahmininizin ya doğru uzunluğa sahip bir kelime ya da harf olması gerekmektedir.")
        counter +=1
        continue
      else:
        for i in self.kelime:
          if i in dogru_harfler:
            print(i,end=" ")
          else:
            print("_",end=" ")




```

```python
oyun1 = adam_asmaca(["metin","kalem","kitap","bilgisayar","bardak"])
```

```python
oyun1.play()
```

```python
print(oyun1)
```

```python
oyun2 = adam_asmaca(["kelime","cümle","makine","bardak","masa"])
```

```python
oyun2.play()
```

```python
kullanici_listesi = {"169854989":{"isim-soyisim":"eren gemici",
                                  "yaş":21}}
#Kullanıcı ekleme fonksiyonu
def kullanici_ekle(tc_no,isim_soyisim,yas):
  kullanici_listesi[tc_no] = {"isim-soyisim":isim_soyisim,"yaş":yas}
  print(f"{isim_soyisim} adlı kullanıcı listeye eklendi")

#Kullanıcı silme fonksiyonu
def kullanici_sil(tc_no):
  if tc_no in kullanici_listesi:
    #del kullanici_listesi[tc_no]
    print(f"{tc_no} numaralı {kullanici_listesi[tc_no]['isim-soyisim']} adlı kullanıcı silindi")
    kullanici_listesi.pop(tc_no)
  else:
    print(f"{tc_no} numaralı kullanıcı bulunamadı")

# Kullanıcı Listele
def kullanici_listele():
  for i in kullanici_listesi.items():
    tc_no,isim_soyisim,yas = i[0],i[1]["isim-soyisim"],i[1]["yaş"]
    print(f"{tc_no},{isim_soyisim},{yas}")

#Giriş Onayla
def giris_onay(tc_no):
  if tc_no in kullanici_listesi:
    if kullanici_listesi[tc_no]["yaş"]>=18:
      print("Giriş onaylandı")
    else:
      print("Yaşınız giriş için yetersizdir")
  else:
    print("Kullanıcı bulunamadı")

```

```python
# class oluşturuyoruz
class ogrenci_kontrol:
  # ilk değişkenlerin tanımlanması için kullanıcıdan aldığımız tcno, isim,soyisim ve yas bilgisini
  # önceden hazırladığımız veritabanına kaydediyoruz
  def __init__(self, tc_no,isim_soyisim,yas):
    self.sozluk = {tc_no:{"isim-soyisim":isim_soyisim,
                          "yaş":yas}}
  # kullanıcıdan aldığımız tcno isim soyisim ve yaş bilgisini veritabanına ekliyoruz.
  def kullanici_ekle(self,tc_no,isim_soyisim,yas):
    self.sozluk[tc_no] = {"isim-soyisim":isim_soyisim,"yaş":yas}
    print(f"{isim_soyisim} adlı kullanıcı listeye eklendi")
  # önce kullanıcı veritabanında kayıtlı mı kontrolünü yapıyoruz
  # sonra da onay metnini yazdırıp kullanıcıyı kaldırıyoruz
  def kullanici_sil(self,tc_no):
    if tc_no in self.sozluk:
      print(f"{self.sozluk[tc_no]['isim-soyisim']} adlı kullanıcı veritabanından kaldırılmıştır.")
      self.sozluk.pop(tc_no)
  # sözlük içerisinde bir döngü oluşturarak kullanıcıların bilgilerini teker teker yazdırıyoruz
  def kullanicilari_listele(self):
    for i in self.sozluk.items():
      tc_no,isim_soyisim,yas = i[0],i[1]["isim-soyisim"],i[1]["yaş"]
      print(f"{tc_no},{isim_soyisim},{yas}")
  # belirli bir tc ye ait yaş bilgilerini kontrol ediyoruz
  def yas_kontrol(self,tc_no):
    if tc_no in self.sozluk:
      if self.sozluk[tc_no]["yaş"]>=18:
        print("Giriş onaylandı")
      else:
        print("Yaşınız giriş için yetersizdir")
    else:
      print("Kullanıcı bulunamadı")
```
