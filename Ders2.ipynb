```python
"""Önce bir sözlük oluşturuyoruz, bu sözlük içerisinde beş tane kullanıcı olsun, bu sözlük öğrenci veritabanı görevi görecek.
Sözlük içeriği, kullanıcı ismi, kullanıcı soyismi, kullanıcı yaşı, kullanıcının sınıfı, kullanıcının notları
Kullanıcıları for döngüsü ile loop a alıyoruz.
Öncelikle, looptaki kullanıcının matematik notlarının ortalaması, 45'ten düşükse kullanıcıya, matematikten kaldınız mesajı yazdırılacak.
Kullanıcı matematikten geçtiyse ise, kullanıcının ismi matematikten geçenler sözlüğüne alınacak.

Kullanıcının türkçe notlarının ortalamasını alıyoruz, ve eğer kullanıcı türkçeden kaldıysa kaldınız mesajı yazdırıyoruz,
Geçtiyse de türkçeden geçenler sözlüğüne alıyoruz

Sonrasında, Türkçeden geçenler sözlüğünü, elemanlarına göre sıralıyoruz.
Sıraladığımız elemanları bir döngüye alıyoruz ve döngüde sırasıyla geçtiniz mesajı yazdırıyoruz.

En son olarak da Matematikten geçenler sözlüğünü elemanlarına göre sıralıyoruz,
Sıraladığımız elemanları bir döngüye alıp geçtiniz mesajı yazdırıyoruz.
"""
```

```python
# Kullanıcılardan oluşan bir sözlük oluşturuyoruz
kullanıcılar = {1:{"isim": "Nevzat",
                   "Soyisim": "Demirci",
                   "yaş": "12",
                   "notlar": {"mat":[15,65,45],
                              "tur":[60,70,80]}},
                2:{"isim": "Mouse",
                   "Soyisim": "Demirci",
                   "yaş": "19",
                   "notlar": {"mat":[79,65,45],
                              "tur":[60,20,10]}},
                3:{"isim": "Klavye",
                "Soyisim": "Demirci",
                "yaş": "18",
                "notlar": {"mat":[67,65,45],
                            "tur":[90,70,80]}},
                4:{"isim": "Çorba",
                "Soyisim": "Demirci",
                "yaş": "17",
                "notlar": {"mat":[15,30,45],
                            "tur":[10,10,80]}},
                5:{"isim": "Fırsat",
                "Soyisim": "Demirci",
                "yaş": "14",
                "notlar": {"mat":[15,65,45],
                            "tur":[60,70,80]}}
                }
```

```python
# matematikten ve türkçeden geçen kullanıcıların notlarının ve isimlerinin depolanması için birer liste oluşturuyoruz
mat_gecen = []
tur_gecen = []

# kullanıcılların bilgilerini index ve değerler olarak iki değişkende alıyorum
for index,item in kullanıcılar.items():

    # matematik notlarını hesaplamak için kullandığım listeyi alıyorum
    mat_not = kullanıcılar[index]["notlar"]["mat"]

    # aldığım listedeki değerlerin toplamını alıp liste uzunluğuna bölerek aritmetik ortalama değerini buluyorum
    mat_not_ort = sum(mat_not)/ len(mat_not)

    # eğer not ortalaması 45 ten büyükse kullanıcıya mesaj yazdırıp listeye ekliyorum
    if mat_not_ort >=45:
        print(f"{kullanıcılar[index]["isim"]} adlı kullanıcının matematik notu {mat_not_ort}'dur. Matematikten geçmiştir.")
        mat_gecen.append(kullanıcılar[index]["isim"]+" "+kullanıcılar[index]["Soyisim"]+"-"+str(mat_not_ort))
        #"Eren GEMİCİ-67"
        #"isim soyisim-ortalama"

    # ortalama 45 ten küçükse kullanıcıya yalnızca mesaj yazdırıyorum
    else:
        print(f"{kullanıcılar[index]["isim"]} adlı kullanıcının matematik notu {mat_not_ort}'dur. Matematikten kalmıştır.")

    #aynı metodu türkçe notları için yapıyorum
    tur_not = kullanıcılar[index]["notlar"]["tur"]
    tur_not_ort =  sum(tur_not) / len(tur_not)
    if tur_not_ort >=45:
        print(f"{kullanıcılar[index]["isim"]} adlı kullanıcının Türkçe notu {tur_not_ort}'dur. Türkçeden geçmiştir.")
        tur_gecen.append(kullanıcılar[index]["isim"]+" "+kullanıcılar[index]["Soyisim"]+"-"+str(tur_not_ort))
    else:
        print(f"{kullanıcılar[index]["isim"]} adlı kullanıcının Türkçe notu {tur_not_ort}'dur. Türkçeden kalmıştır.")
```

```python
# ["Eren GEMİCİ","65"]
mat_gecen = sorted(mat_gecen,key = lambda i:(i.split("-")[1]),reverse=True)
tur_gecen = sorted(tur_gecen,key = lambda i:(i.split("-")[1]),reverse=True)
for index in range(len(mat_gecen)):
    isim,notu = mat_gecen[index].split("-")
    print(f"{isim} adlı kullanıcı sınıfta matematikten {notu} puanı ile geçen ve Matematikte {index+1}. sırada olan öğrencidir")

for index in range(len(tur_gecen)):
    isim,notu = tur_gecen[index].split("-")
    print(f"{isim} adlı kullanıcı sınıfta matematikten {notu} puanı ile geçen ve Türkçede {index+1}. sırada olan öğrencidir")


```
