 # String metotları

```python
isim = "Eren gemici"
print(isim.upper()) # string içerisindeki değerlerin hepsini büyük harf yapar.
print(isim.lower()) #string içerisindeki değerlerin hepsini küçük harf yapar.
print(isim.capitalize())#Sadece baştaki harfi büyütür.
print(isim.title())# Tüm kelimelerin baş harflerini büyütür.
print(isim.swapcase()) # Küçükse büyük, büyükse küçük harf yapar.

isim = "    Eren    "
print(isim)
print(isim.strip())
print(isim.lstrip())
print(isim.rstrip())

harfler = "evin çitini yeniliyoruz"
# bir değişiklik yaptığımızda onu değişkenin kendisine tekrar atamamız gerekir
harfler = harfler.replace("çit","tip")
print(harfler)
```
